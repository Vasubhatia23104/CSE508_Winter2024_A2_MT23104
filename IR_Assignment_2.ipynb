```python
import pandas as pd
import math
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter
import numpy as np

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

all_image_review=[]
# Function to preprocess text
def preprocess_text(text):
    # Ensure text is a string
    text = str(text)
    # Remove HTML tags
    clean_text = re.sub('<.*?>', '', text)
    # Convert to lowercase
    clean_text = clean_text.lower()
    # Tokenize
    tokens = word_tokenize(clean_text)
    # Remove stopwords and non-alphabetic tokens
    stop_words = set(stopwords.words('english'))
    words = [word for word in tokens if word.isalpha() and word not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    return words

# Function to calculate term frequency (TF)
def calculate_tf(words):
    word_count = len(words)
    term_frequency = {}
    word_freq = Counter(words)
    for word in word_freq:
        term_frequency[word] = word_freq[word] / word_count
    return term_frequency

# Function to calculate inverse document frequency (IDF)
def calculate_idf(data):
    total_documents = len(data)
    idf = {}
    for index, row in data.iterrows():
        review = row['Review Text']
        words = set(preprocess_text(review))
        for word in words:
            idf[word] = idf.get(word, 0) + 1
    for word in idf:
        idf[word] = math.log(total_documents / (idf[word] + 1))
    return idf

# Calculate TF-IDF scores for each review (whole sentence)
def calculate_tf_idf(data):
    index=0
    tf_idf_scores = {}
    idf = calculate_idf(data)
    for i, row in data.iterrows():
        review = row['Review Text']
        # print(review)
        words = preprocess_text(review)
        tf = calculate_tf(words)
        tf_idf = {}
        for word in words:
            tf_idf[word] = tf[word] * idf.get(word, 0)
        # Concatenate all TF-IDF scores into a single vector
        tf_idf_vector = np.array(list(tf_idf.values()))
        image_urls = eval(row['Image'])
        for i in  image_urls :
           all_image_review.append(review)
           tf_idf_scores[index] = tf_idf_vector
           index=index+1
    return tf_idf_scores

csv_file = "/content/drive/MyDrive/Ir_Assignment2.csv"

# # Read CSV file
df = pd.read_csv(csv_file)
#df = pd.DataFrame(data)

# Calculate TF-IDF scores for the dataframe
tf_idf_review = calculate_tf_idf(df)
print(tf_idf_review)
# Example usage
# for review, scores in tf_idf_dict.items():
#     print("Review:", review)
#     print("TF-IDF Scores:", scores)
```
